Remember what Wes McKinney said about tidy data?

```python
#pd.melt, arguments: 'London_Borough','ID'
clean_properties = pd.melt(properties_T, id_vars= ['London_Borough', 'ID'])
clean_properties.head()

# Re-name the column names: Rename the '0' column 'Month', and the 'value' column 'Average_price';.rename()
clean_properties = clean_properties.rename(columns = {0:'Month','value':'Average_price'})
clean_properties.head()

#use the .dtypes attribute to check the data types of our clean_properties DataFrame:
clean_properties.dtypes
#objects, want numeric value for Average_price, not object

```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
#Call the to_numeric() method on pd, pass the 'Average_price' column into its brackets, and reassign the result to the clean_properties 'Average_price' column.
clean_properties['Average_price'] =  pd.to_numeric(clean_properties['Average_price'])
clean_properties.dtypes  #Average_price now a float, ok?

#Review dataframe: .count() columns with values
clean_properties.count() #not all columns has same count of values... must be missing data, or NaN values?
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
#Why are there not equal count observations? .unique()
clean_properties['London_Borough'].unique() # some weird names, can we elimate these? 'Unnamed: 47' for example

#Determine how many are not actual boroughs: subsetting
clean_properties[clean_properties['London_Borough']=='Unnamed: 37'].head()
clean_properties[clean_properties['London_Borough']=='Unnamed: 47'].head()
clean_properties[clean_properties['ID'].isna()]  #'ID' == NaN

#Remove, delete these rows from above, since they don't have relevant data; 'ID' or 'Average_price'==NaN; .notna()
NaNFreeDF1 = clean_properties[clean_properties['Average_price'].notna()]
NaNFreeDF1.head(50)
NaNFreeDF1.count() #all columns have matching values now, great!

#Remove non-Borough values under 'London_Borough':
NaNFreeDF1['London_Borough'].unique()                           #extract unique str values in 'London_Borough'
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']
#Filter out the NotBorough under 'London_Borough'; ~ operating before filter statement to remove 
NaNFreeDF1[~NaNFreeDF1.London_Borough.isin(nonBoroughs)]
NaNFreeDF1.head(25)  #looks like nonBoroughs were removed
NaNFreeDF1.count()     #counts like all matched up re: column values

#rename NaNFreeDF1 to 'df'
df= NaNFreeDF1
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#create function 'create_price_ratio'
def create_price_ratio(d):
    y1998 = float(d['Average_price'][d['year']==1998])
    y2018 = float(d['Average_price'][d['year']==2008])
    ratio = [y2018/y1998]    #should there be a greater than 1 ratio for growth?
    return ratio
#plug in provided example
create_price_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham'])

#make an empty dictionary, called final, where we'll store our ratios for each unique London_Borough.
final = {}

#declare loop to iterate through each unique element of 'London_Borough' in dfg:
for x in dfg['London_Borough'].unique():
    #make parameter to our create_price_ratio function, assign to new varibale: 'borough'
    borough = dfg[dfg['London_Borough']== x]
    # Make a new entry in the final dictionary whose value's the result of calling create_price_ratio with the argument: borough
    final[x] = create_price_ratio(borough)
    
#print(final)

#convert final to a dataframe; assigned to new variable: df_ratio
df_ratio = pd.DataFrame(final)

#transpose df_ratio: .T
df_ratio_T = df_ratio.T

#Resert index of df_ratio_T: .reset_index()
df_ratio_T = df_ratio_T.reset_index()

df_ratio_T.head() 

#rename the 'index' column as 'London_Borough', and the '0' column to '2018'.
df_ratio_T.rename(columns={'index':'Borough',0:'2018'},inplace=True)
df_ratio_T.head()

#Want to find which London Boroughs have increase in housing price, avergage, over this period
#Let's look at all Boroughs, sort from highest to lowest (descending order)
df_LondonBoroughPrices = df_ratio_T.sort_values(by='2018',ascending=False).head(15)
df_LondonBoroughPrices.head(10)

#Visualize with plot: bar/histogram? 
plot_df_prices = df_LondonBoroughPrices[['2018','Borough']].plot(kind='bar')  #df[#].plot(kind='#')
plot_df_prices.set_xticklabels(df_LondonBoroughPrices.Borough)     #set_xticklabels()
plot_df_prices.set_ylabel('Ratio of Change from 1998 to 2018')
plot_df_prices.set_xlabel('Borough of London')
```
